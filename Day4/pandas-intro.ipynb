```python
import pandas as pd

print('Version: ',pd.__version__)
```

```python
list = [1,0,4,29,54,65,20]
mySeries = pd.Series(list)
print(mySeries)
print(f"Element at index 3 is: {mySeries[3]}")

a = [1,5,6]
mySeries = pd.Series(a,index=["a","b","c"])
print(mySeries)
print(mySeries["b"])
```

```python
employee = {
    "id":1,
    "name":"AA",
    "email":"aa@gmail.com",
    "address":"US",
    "salary":2345
}

data = pd.Series(employee)
print(data)
print(f"id {data["id"]} ")
```

```python
sample={
    "id":[1,2,3,4],
    "name":["AA","BB","CC","DD"]
}

data = pd.DataFrame(sample)
print(data)


data = {
    "Name":["AA","BB","CC"],
    "Age":[20,23,19],
    "City":["US","UK","IN"]
}

data1 = pd.DataFrame(data)
print(data1)

print(data1.loc[0])
print(data1.loc[[0,1]])

print(data1.iloc[0])

df1 = pd.DataFrame(data,columns=["Name", "City", "Age"])
print(df1)
```
