```python
import matplotlib.pyplot as plt

print(plt.__package__)
```

```python
plt.figure(figsize=(6,4))
plt.plot([1,3,5],[2,4,6])
plt.show()
```

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]

plt.plot(x,y,marker='o',linestyle='-',color='b')

plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Simple Line Plot")
plt.grid(True)
plt.show()

```

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]

plt.scatter(x,y,color='r',marker='*')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

```python
categories = ['A','B','C','D']
values = [5,7,3,8]

5plt.bar(categories,values,color='g')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar Chart")
plt.show()
```

```python
import numpy as np

data = np.random.randn(1000)
plt.hist(data,bins=30,color='purple',edgecolor='black')
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
sizes = [20,30,25,25]
labels = ['JS','Java','React','Angular']
colors=['blue','red','orange','green']
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    "x":[1,2,3,4,5],
    "y":[10,20,30,40,50]
})

sns.lineplot(x="x",y="y",data=data)
plt.title("Seaborn Line Plot")
plt.show()
```
