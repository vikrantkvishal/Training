```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Assignment Data.csv')

# Task 1 : Display Basic Data
print(data.head(20))
```

```python
# Task 2 : Average salary in each department

avg_Data = data.groupby('Department')['Salary'].mean()
print(avg_Data)
```

```python
# Task 3 : Highest and Lowest Salary in Company

highest = data['Salary'].max()
lowest = data['Salary'].min()

print('Highest Salary :',highest)
print('Lowest Salary :',lowest)
```

```python
# Task 4 : Employee with Maximum Salary

max_Salary = data[data['Salary'] == highest]
print('Employee with Maximum Salary :')
print(max_Salary)
```

```python
# Task 5 : Count employees in each department

count_employees = data.groupby('Department')['Employee_ID'].count()
print('Count employees in each department :')
print(count_employees)
```

```python
# Task 6 : Bar Chart for number of employees in each department

plt.figure(figsize = (10,8))
count_employees = data.groupby('Department')['Employee_ID'].count().plot(kind = "bar" , color = 'green')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Number of Employees in each Department')
plt.grid(linestyle = "--", alpha = 0.5)
plt.show() 
```

```python
# Task 7 : Pie Chart Salary Distribution by Department

plt.figure(figsize = (10,8))
data.groupby('Department')['Salary'].sum().plot(kind = "pie" , autopct = '%1.1f%%')
plt.title('Scatter Plot -- Salary vs Experience')
plt.ylabel("")
plt.show()
```

```python
# Task 8 : Scatter Plot Salary vs Experience

plt.figure(figsize = (10,8))
plt.scatter(data['Salary'], data['Experience(in Years)'], color = 'blue')
plt.xlabel('Salary')
plt.ylabel('Experience(in Years)')
plt.title('Scatter Plot -- Salary vs Experience')
plt.show()
```

```python
# Task 9 : Employee with salary more than average salary
avg = data['Salary'].mean()
Employee_with_more_average_salary = data[data['Salary'] > avg]
print('Employee with salary more than average salary :')
print(Employee_with_more_average_salary)
```

```python
# Task 10 : Sort Employees based on Salary in Descending Order

sorted_Data = data.sort_values(by='Salary', ascending=False)

print("Sorted Salary (Descending Order) :")
print(sorted_Data)
```
